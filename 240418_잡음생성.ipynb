잡음 생ㅅ엉

```python
import numpy as np
import matplotlib.pyplot as plt

# 잡음의 전력 설정
noise_power = 6.31 # For -8dB. If -22dB, set noise_power to 158.49

# 잡음의 크기 설정
num_samples = 1023
sigma = np.sqrt(noise_power)

# 실수부와 허수부에 대한 잡음 생성
real_noise = np.random.normal(0, sigma, num_samples)
imaginary_noise = np.random.normal(0, sigma, num_samples)

# 각각의 전력 계산
real_power = np.mean(real_noise ** 2)
imaginary_power = np.mean(imaginary_noise ** 2)

print("실수부 전력:", real_power)
print("허수부 전력:", imaginary_power)

# 복소수로 변환
complex_noise = (real_noise + 1j * imaginary_noise) / np.sqrt(2)

# 복소평면에 잡음 플로팅
plt.figure(figsize=(8, 6))
plt.scatter(np.real(complex_noise), np.imag(complex_noise), s=5)
plt.title("Complex Noise in the Complex Plane")
plt.xlabel("Real Part")
plt.ylabel("Imaginary Part")
plt.grid(True)
plt.show()
```
